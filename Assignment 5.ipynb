```python
#1 Given a list of integers, write a function to return the sum of all prime numbers in that list.
import math
def prime(n):
    if n == 0 or n == 1:
        return False
    else:
        for i in range(2, int(n**0.5)+1):
            if n%i == 0:
                return False
                break
        else:
            return True
            
def sum_prime(ls):
    sum_of_primes=0
    print("Prime numbers in list :")
    for i in ls:
        if prime(i):
            print(i)
            sum_of_primes += i
    print("Sum of primes :",sum_of_primes)
#sum_prime([0,1,2,3,4,5,10,11,12,13])
n = int(input('Enter the length of list '))
print('Enter the elements of list:')
l = []
for i in range(0,n):
    x = int(input())
    l.append(x)
sum_prime(l)
print()
```

```python
#Given a list of integers, write a function to check whether the list is strictly increasing or not.
def Inc(l):
    flag=True
    for i in range(1,len(l)):
        if l[i-1]>l[i]:
            flag=False
            break
    return flag
l1=input().split()
list1=list(map(lambda x:int(x),l1))
print('List is strictly increases : ',Inc(list1))
```

```python
#Write a function to check whether a given list is expanding or not (the difference between adjacent elements should keep on increasing).
def expanding(l):
    d1 = abs(l[1] - l[0])
    d2 = abs(l[2] - l[1])
    
    if d1 >= d2:
        return False
    
    else:
        for i in range(3,len(l)):
            d3 = abs(l[i] - l[i-1])
            if d3 <= d2:
                return False
            d2 = d3
        else:
            return True
expanding([1,3,7,2,9])
```

```python
#Write a function to calculate all permutations of a given string (without using itertools).
def permutation(str) :
    if len(str) == 1 :
        return [str]
    prmts = permutation(str[1:])
    fst = str[0]
    rslt = []
    for i in prmts :
        for j in range(len(i)+1) :
            rslt.append(i[:j] + fst + i[j:])
    return rslt


inp = str(input("Enter the String : "))
permutation(inp)
```
